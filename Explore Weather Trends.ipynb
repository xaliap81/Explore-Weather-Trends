## Explore Weather Trends

### Importing necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Load the data

```python
global_temp = pd.read_csv("Global.csv")
london = pd.read_csv("London.csv")
berlin = pd.read_csv("Berlin.csv")
new_york = pd.read_csv("NY.csv")
```

### Visualizing the cities average temperature with average global

```python
plt.figure(figsize=(14,12))

# London Average Temperature vs. Global
plt.subplot(3,1,1)
plt.plot(global_temp['year'],global_temp['avg_temp'],label='Global',color="darkblue",linestyle="--")
plt.plot(london['year'],london['avg_temp'],label='London', color="orchid",marker=",")
plt.legend(prop={'size':14},loc="lower right")
plt.xticks(range(1800,2021,10))
plt.xlabel("Years")
plt.ylabel("Temperature in Celsius") 
plt.title("London City Average Temperature vs Global Temperature")

# Berlin Average Temperature vs. Global
plt.subplot(3,1,2)
plt.subplots_adjust(top=1.5)
plt.plot(global_temp['year'],global_temp['avg_temp'],label="Global",color="darkblue",linestyle="--")
plt.plot(berlin['year'],berlin['avg_temp'],label="Berlin",color="tomato")
plt.legend(prop={'size':14},loc="lower right")
plt.xticks(range(1800,2021,10))
plt.xlabel("Years")
plt.ylabel("Temperature in Celsius")
plt.title("Berlin City Average Temperature vs Global Temperature")

# New York Average Temperature vs. Global
plt.subplot(3,1,3)
plt.subplots_adjust(top=1.5)
plt.plot(global_temp['year'],global_temp['avg_temp'],label="Global",color="darkblue",linestyle="--")
plt.plot(new_york['year'],new_york['avg_temp'],label="New York", color="yellowgreen",marker=",")
plt.legend(prop={'size':14},loc="lower right")
plt.xticks(range(1800,2021,10))
plt.xlabel("Years")
plt.ylabel("Temperature in Celsius")
plt.title("New York City Average Temperature vs Global Temperature")
          
plt.show()
```

### Calculating the moving average for three cities and global data

```python
global_mv_avg = global_temp['avg_temp'].rolling(10).mean()
london_mv_avg = london['avg_temp'].rolling(10).mean()
berlin_mv_avg = berlin['avg_temp'].rolling(10).mean()
new_york_mv_avg = new_york['avg_temp'].rolling(10).mean()
```

#### Exploring the moving averages

```python
global_mv_avg.tail()
```

```python
new_york_mv_avg.tail()
```

### Visualizing the cities moving average vs. global moving average

```python
plt.figure(figsize=(14,12))
plt.rc('grid',linestyle="--",color="silver")

# London Moning Average Temperature vs. Global
plt.subplot(3,1,1)
plt.plot(global_temp['year'],global_mv_avg,label='Global',color="navy")
plt.plot(london['year'],london_mv_avg,label='London', color="plum")
plt.legend(prop={'size':14},loc="lower right")
plt.grid(True)
plt.xticks(range(1800,2015,10))
plt.xlabel("Years")
plt.ylabel("Temperature in Celsius") 
plt.title("London City Moving Average Temperature vs Global Moving Average Temperature")

# Berlin Moving Average Temperature vs. Global
plt.subplot(3,1,2)
plt.subplots_adjust(top=1.5)
plt.plot(global_temp['year'],global_mv_avg,label="Global",color="navy")
plt.plot(berlin['year'],berlin_mv_avg,label="Berlin",color="red")
plt.legend(prop={'size':14},loc="lower right")
plt.grid(True)
plt.xticks(range(1800,2015,10))
plt.xlabel("Years")
plt.ylabel("Temperature in Celsius")
plt.title("Berlin City Moving Average Temperature vs Global Moving Average Temperature")

# New York Moving Average Temperature vs. Global
plt.subplot(3,1,3)
plt.subplots_adjust(top=1.5)
plt.plot(global_temp['year'],global_mv_avg,label="Global",color="navy")
plt.plot(new_york['year'],new_york_mv_avg,label="New York", color="green")
plt.legend(prop={'size':14},loc="lower right")
plt.grid(True)
plt.xticks(range(1800,2015,10))
plt.xlabel("Years")
plt.ylabel("Temperature in Celsius")
plt.title("New York City Moving Average Temperature vs Global Moving Average Temperature")

plt.show()
```
